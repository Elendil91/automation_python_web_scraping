Extract CSV Files from Websites

```python
#reading 1 cvs file from the website
#target website: https://www.football-data.co.uk/englandm.php

df_england = pd.read_csv('https://www.football-data.co.uk/mmz4281/2122/E0.csv')
```

```python
#rename columns

df_england.rename(columns={'FTHG': 'home_goals',
                           'FTAG': 'away_goals'}, inplace=True) 
```

```python
import camelot 

tables = camelot.read_pdf('datacleaning.pdf', pages='9') #parameter('file.pdf', pages= '1')

print(tables)
```

```python
tables.export('datacleaning.csv', f='csv', compress=True) #parameter('file_with_new_format.cvs', f='format', compress=True)
tables[0].to_csv('datacleaning.csv') #export new file
```
